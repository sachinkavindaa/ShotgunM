```python
import pandas as pd

# Read the mappingfile.csv into a DataFrame
df = pd.read_csv('mappingfile.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure sample_date is parsed as a date if not already
df['sample_date'] = pd.to_datetime(df['sample_date'], errors='coerce')

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='sample_date', y='animal_id', hue='ibk_status', palette='Set1')
plt.title('Animal ID, Sample Date, and IBK Status')
plt.xlabel('Sample Date')
plt.ylabel('Animal ID')
plt.legend(title='IBK Status')
plt.tight_layout()
plt.show()
```
